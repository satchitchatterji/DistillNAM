```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
# Load the CIFAR-10 dataset
mnist = keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Preprocess the data
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
train_images = train_images.reshape(-1, 28*28)
test_images = test_images.reshape(-1, 28*28)
train_images.shape
```

```python
def single_model(input_layer):
#     dense1 = keras.layers.Dense(1, activation='relu')(input_layer)
#     dense2 = keras.layers.Dense(16, activation='relu')(dense1)
    dense3 = keras.layers.Dense(1, activation='linear')(input_layer)
    return dense3

inputs = [keras.layers.Input(shape=(1)) for _ in range(784)]
models = [single_model(inputs[i]) for i in range(784)]

models = keras.layers.concatenate(models)
output = keras.layers.Dense(10)(models)

model = keras.Model(inputs=inputs, outputs=[output])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
train_list = train_images.reshape(784,-1)
```

```python
p = [x for x in train_list]
```

```python
# Train the model
history = model.fit([x for x in train_list], np.array(train_labels), epochs=20, batch_size=32,verbose=1)

# Get the training and validation loss values from the history object
train_loss = history.history['loss']
val_loss = history.history['val_loss']

# Get the training and validation accuracy values from the history object
train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

# Plot the training and validation losses
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(range(len(train_loss)), train_loss, label='Training Loss')
plt.plot(range(len(val_loss)), val_loss, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()

# Plot the training and validation accuracies
plt.subplot(1, 2, 2)
plt.plot(range(len(train_acc)), train_acc, label='Training Accuracy')
plt.plot(range(len(val_acc)), val_acc, label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
